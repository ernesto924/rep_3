prendere la lista di coefficienti e un punto es x=3 e restituisce il valore del polinomio

```python

def fun_1(a:list, b:int ):
    r=0
    for i in range(len(a)):
        r += a[i]*b
    return r
```

```python

a=[1,2,3]
b=3


fun_1(a,b,)
```

prendere una lista di coefficienti

```python
def fun_2(a:list, b:list ):
    r=0
    for i in range(len(a)):
        for j in range(len(b)):
            r += a[i]*b[j]
    return r
```

```python
a=[1,2,3]
b=[1,2,3]
fun_2(a,b)
```
