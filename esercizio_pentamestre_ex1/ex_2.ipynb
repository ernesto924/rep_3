### Ex 2: Scrivi una funzione che prende in input una matrice e restituisce il numero di volte che compare la cifra 2

```python
import numpy as np
A=np.array([[2,2],[3,2]])
n_2=0
for i in range(A.shape[0]):
    for j in range(A.shape[0]):
        if A[i][j]==2:
            n_2=n_2+1
print(n_2)
```
