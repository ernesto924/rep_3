### Ex 1 [1.25 pt]: Elemento minimo in una lista  

Scrivi una funzione che prende in input una lista e restituisce l'elemento più piccolo

Es: Input: [1, 3, -1, 3, 4] -> Output: -1

````python
def fun_1(a:list) -> int:
    min=a[0]
    for i in range(len(a)):
        if a[i]<min:
            min = a[i]
    return min
````

````python
a=[1, 3, -1, 3, 4]
min=fun_1(a)
print(min)
````

### Ex 2 [1.25 pt]: Somma multipli di 5

Scrivi una funzione che restituisce la somma degli elementi divisibili per 5

Es: Input: [1, 10, -1, -5] -> Output: 5

````python
def fun_2(a:list) -> int:
    somma=0
    for i in range(len(a)):
        if a[i] % 5 == 0:
            somma += a[i]
    return somma
````

````python
a=[1, 10, -1, -5]
somma=fun_2(a)
print(somma)
````

### Ex 3 [1.5 pt]: Conteggio della frequenza di un certo elemento

Scrivi una funzione che prende in input una lista di numeri ed un valore. Restituisce il numero di volte che tale valore compare nella lista.

Esempio: Input: [1, 2, 3, 2], 3 -> Output: 1

````python
def fun_3(a:list, b:int) -> int:
    conto=0
    for i in range(len(a)):
        if a[i] == b:
            conto += 1
    return conto
````

````python
conto=fun_3([1, 2, 3, 2], 3)
print(conto)
````

````python
import numpy as np

def fun_6(a:np.ndarray, x:list, y:list, n:int ) -> int:
    r=0
    for i in range(n):
        for j in range(n):
            r += x[i]*a[i][j]*y[j]
    return(r)
    
````
