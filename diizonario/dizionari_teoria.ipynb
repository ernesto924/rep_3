```python
frase="Ciao! come va? sto bene grazie"
d={}
for c in frase:
    if c in d:
        d[c] +=1
    else:
        d[c]=1
for k in d.keys():
    print(k, d[k])
#modo 2 di stampare 
```

```python
#esempio
def fun_1(a:str) -> dict:
    d={}
    for c in a:
        if c in d:
            d[c]+=1
        else:
            d[c]=1
    return d
```

```python
A="a,b,a,b"
var=fun_1(A)
print(var)
```
